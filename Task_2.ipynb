What are possible bioinformatics and, more specific, systems biology questions you could identify wrt.
the COVID-19?

1. Bioinformatic related
- How can we simulate the relation between oxygen and bacteria which may or may not lead to infection?
- Is there a reliable approximation about how many people an individual infects within the critical span of 20 days?
- How do we gather data about the effectiveness of infection-stopping-strategies, such as vaccines, isolation, masks, hygiene, ..?
- What are the ethics behind using test results for scientific research or making testing an aobligation in general from a perspective where the data is crucial for progress?
- As scientists, how do we make sure that for each problem we collectively use similiar and correct numbers for research but also media communication (e.g. CFR, IFR)?

2. Systems Biology related
- Could symptoms, their first appearances, intensity, and duration, which for COVID vary a lot, be predicted based on an individual's clinical picture?
- Can systems biology help us to come up with more reliable tests and vaccines?
- How reliable can bacteria from animals, in terms of "zoonosis", be identified within humans' immune systems?
- Is there a way of blocking ACE2 rezeptors endangered to be accessed by the viruse's "spikes"?
- In order to protect the body as a whole, where is be the best location to place monoclonal antibodies if accurat placement is possible?
- Can system biology help in eliminating generic risk factors, such as substituting type I interferons or produce auto-antibodies?
- Why are some animals (not) immune to COVID? Can comparsions between our and their systems help us in research?
