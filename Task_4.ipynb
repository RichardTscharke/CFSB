**General information** (CS-only student)

- An organism's complete set of **DNA** (contains the instructions needed to develop and direct the activities of a living organism) is called its **genome**.
- A **gene** is the unit of DNA that carries the instructions for making a specific (set of) protein(s). This process is called **Transcription**.
- An **enzyme** (RNA polymerase) copies the information in a gene's DNA into the so-called **mRNA**.
- The mRNA is read by a **ribosome** and the information is used to form a specific **protein**. This process is called **Translation**.
- Proteins make up body structures (e.g. organs and tissue), control chemical reactions and carry signals between cells.

[1]

**RNA-viruses**
- **Viruses** are small unicellular microorganisms which contain only one type of nucleic acid (DNA or RNA).
- The **RNA viruses** provide the only instance in nature in which RNA is the exclusive repository of genetic information.

[2]

Consequently, since viruses like SARS-CoV-2 store their genetic info directly as RNA, the Transcription process is skipped. The RNA can be read directly by the ribosomes, making it immediately infectious upon entering a cell.

# Collect genomic and proteomic information about the SARS-CoV-2 virus.

**Genomic information**
- **GenBank accession number:** NC_045512
- **Genome type:** single-stranded RNA
- **Length:** ~30,000 bases
- **Sense/Polarity:** positive (The host cell's ribosome can immediately read it and produce proteins without the conversion to mRNA)

**Proteomic information**
- **#Proteins:** 29 (16 non-structural, 4 structural, 9 accessory)
- **Emergence:** The ribosome translates ORF1ab into one large "polyprotein", which is then cleaved into the 16 NSPs by viral proteases (nsp3, nsp5).
- **Non-structural:** refer to Table 1 of [3] for a description of Protein-ID, PDB-Code, and functions.
- **Structural:** 
    1. spike (S): recognizes the host receptor to mediate the virus into the host cells. S protein are the main target of many anti-coronavirus drugs.
    2. membrane (M): Maintains the virion's size and shape, involved in processing, modification, and trafficking of multiple viral proteins, assembly, release of viruses.
    3. envelope (E): Assists in assembly and performs "viral budding" (exiting host cells without immediately destroying them => production, stealth,...) 
    4. nucleocapsid (N): Participates in replication, assembly, and suppresses/disrupts a cells alarm systems ("immune evasion").
    
    [4]  

- **Accessory:** perhaps the least explored proteins of SARS-CoV-2 but mentioned functions are:
    1. Immune evasion mechanisms (e.g. "inhibition of cytokine secretion" by ORF9c)
    2. Altering of the host's cellular mechanisms (e.g. "autophagy" or "apoptosis" by ORF9b)
    3. Regulation of viral protein expression via "subgenomic RNAs" (sgRNAs).

    [5]

# Collect information about the SARS-CoV-2 life cycle, in particular the interaction with the host(s).

- **Life cycle:**

    1. **Entry:** The Spike (structural protein S) binds to the host cell's ACE2 receptor enabling so-called **membran fusion**.
    2. **Translation:** The ribosome reads the RNA, builds the polyportein and cleaves it into the 16 NSPs.
    3. **Replication:** The non-structural proteins copy the viral RNA.
    4. **Assembly:** The structual proteins come together to form new virus particles.
    5. **Building/Release:** The structural E protein mediates budding (see above) and new virions exit the cell

- **Host related:**

    - The **transmission** primarily takes place via respiratory droplets and aerosols.
    - The **incubation period** (time between infection and first symptoms) is ~5 days for COVID in which the host can already be infectious.
    - In comparision with other viruses this is relatively short caused mainly by SARS-CoV-2 being positive-sensed (see above)
    - The **innate immunity response** to the virus is initiated by "Pattern Recognition Receptors" which detect viral RNA and proteins. This process can be delayed through viral structural proteins (N).
    - The **adaptive immunity response** invlolves killing detected cells, producing antibodies, and targeting the structural (S) and (N) proteins.
    - SARS-CoV-2 causes **tissue tropism** (virus replicates in a specific tissue or organ) not only in the respiratory system (e.g., lungs) but also heart, kidneys, blood vessels and other tissues. Thus, being a **multiorgan affecting virus**. [6]

- **Evolutionary information:**
    - The virus has a zoonotic origin (jumping from animals, here likely bats, to humans). Wikipedia also suggests that "the emergence of SARS-CoV-2 may have resulted from recombination events between a bat SARS-like coronavirus and a pangolin coronavirus through cross-species transmission." [7]
    - Prior to SARS-CoV-2 in ~2002 a predecessor virus with 79% genomic sequence identity, called SARS-CoV-1, caused ~8,000 cases. But due to a higher mortality and lower transmissible rate, the virus did not become a pandemic. [8] 
    - The WHO classifies variants as **Variants of Concern (VoC)** . The major VoCs were Alpha, Beta, Gamma, Delta, and Omicron — each with distinct mutations primarily in the Spike protein. [9]
